```python
def cars(x):
    if x>10280 and x<30000:
        return "low_Budget"
    elif x>30001 and x<=55000:
        return "mid_Budget"
    elif x>55001 and x<=80000:
        return "High_budget"
    else:
        return "sports cars"
```

```python
df["category"]=df.MSRP.apply(cars)
```
